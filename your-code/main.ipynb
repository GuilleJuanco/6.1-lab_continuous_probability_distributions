## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
import numpy as np
import matplotlib.pyplot as plt

def generate_uniform_random(bottom, ceiling, count):
    return np.random.uniform(bottom, ceiling, count)

set1 = generate_uniform_random(10, 15, 100)

set2 = generate_uniform_random(10, 60, 1000)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(set1, bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('SET1')

plt.subplot(1, 2, 2)
plt.hist(set2, bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('SET2')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# your answer below
#La diferencia reside en el número de valores dentro de la distribución, el segundo plot recoge muchos mas valores dentro de la distribución, siendo distribuidos de manera uniforme en ambos.
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# your code here

def generate_normal_random(average, std_dev, count):
    return np.random.normal(average, std_dev, count)

set1 = generate_normal_random(10, 1, 1000)

set2 = generate_normal_random(10, 50, 1000)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(set1, bins=30, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Normal Distribution (Set 1)')

plt.subplot(1, 2, 2)
plt.hist(set2, bins=30, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Normal Distribution (Set 2)')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# your answer below
#Ambas son distribuciones normales con media 10, la diferencia reside en la desviación típica de ambas, siendo 1 y 50, lo que distribuye de manera mas amplia la segunda, repartiendo los valores entre una gama mas amplia (abanico de 50 a 1).
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
def generate_exponential_random(mean, size):
    return np.random.exponential(mean, size)

Set1 = generate_exponential_random(10, 10)

Set2 = generate_exponential_random(10, 100)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(Set1, bins=100, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('SET1')

plt.subplot(1, 2, 2)
plt.hist(Set2, bins=100, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('SET2')

plt.tight_layout()
plt.show()
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your answer here
# Hint: This is same as saying P(x<15)
from scipy.stats import expon

mean = 10

rate = 1 / mean

prob_15 = expon.cdf(15, scale=1/rate)

print("La probabilidad de pasar menos de 15 minutos en el banco es ", prob_15)
```

What is the probability that the customer will spend more than 15 minutes

```python
# your answer here
print("La probabilidad de pasar mas de 15 minutos en el banco es ", 1-prob_15)
```
